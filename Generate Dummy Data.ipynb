```python
from faker import Faker
import types

fake = Faker()
attributes = dir(fake)

methods = []
for attr in attributes:
    try:
        if isinstance(getattr(fake, attr), types.MethodType):
            methods.append(attr)
    except TypeError:
        continue
for method in methods:
    print(method)
```

```python
import pandas as pd
from faker import Faker

fake = Faker()

num_rows = 100
df = pd.DataFrame({
    'name': [fake.name() for _ in range(num_rows)],
    'address': [fake.address().replace('\n', ', ') for _ in range(num_rows)],
    'email': [fake.email() for _ in range(num_rows)],
    'date_of_birth': [fake.date_of_birth() for _ in range(num_rows)],
})

df
```
