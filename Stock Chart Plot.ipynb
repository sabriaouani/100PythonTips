```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download historical data for S&P 500 index from the last 5 years
data = yf.download('^GSPC', start='2018-08-01', end='2023-08-01')

# Plot the closing prices
data['Close'].plot()
plt.title('S&P 500 Index - Last 5 Years')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()
```
