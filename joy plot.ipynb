```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Let's create a sample dataframe
data = pd.DataFrame({
    'Group': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Value': [1, 2, 3, 2, 3, 1, 3, 1, 2]
})

# Create a figure and axes
fig, ax = plt.subplots()

# We'll use seaborn's kdeplot for each group
for name, group in data.groupby('Group'):
    sns.kdeplot(group['Value'], ax=ax, label=name, shade=True)

plt.show()
```
